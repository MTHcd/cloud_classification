```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers, models

# 1. Préparation des images (Lecture de vos 10 dossiers)
train_datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)

train_generator = train_datagen.flow_from_directory(
    'data/',  # Chemin vers le dossier contenant vos dossiers A, B, C...
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    subset='training'
)

# 2. Structure de l'IA (Modèle simple pour commencer)
model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(train_generator.num_classes, activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# 3. L'ENTRAÎNEMENT (C'est ici que le .h5 se "fabrique")
# Lancez au moins 3 à 5 époques
model.fit(train_generator, epochs=5)

# 4. SAUVEGARDE FINALE
model.save('modele_nuages.h5')
print("Succès ! Le fichier modele_nuages.h5 a été créé dans votre dossier.")
```

```python
import json
import os

# Liste des dossiers dans 'data' triés par ordre alphabétique
data_dir = 'data'
folders = sorted([f for f in os.listdir(data_dir) if os.path.isdir(os.path.join(data_dir, f))])

# Création du dictionnaire {index: nom}
class_indices = {str(i): name for i, name in enumerate(folders)}

# Sauvegarde
with open('classes.json', 'w') as f:
    json.dump(class_indices, f)

print("Fichier classes.json créé avec les classes suivantes :", folders)
```
